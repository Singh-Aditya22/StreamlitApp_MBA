```python
df = raw_data.loc[:,['InvoiceNo', 'StockCode', 'Description']]
df.shape
```

```python
df.dropna(axis=0, inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
basket = df.groupby(['InvoiceNo', 'Description'])['StockCode'].count().unstack().fillna(0)

# Convert counts to 1/0 (did the customer buy it or not)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
```

```python
basket.head()
```

```python
import mlxtend
print(mlxtend.__version__)
```

```python
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
```

```python
# Normalize consequents (convert sets to frozen sets so unordered = same)
rules['consequents'] = rules['consequents'].apply(lambda x: frozenset(x))
rules['antecedents'] = rules['antecedents'].apply(lambda x: frozenset(x))

# Sort by lift to find strong rules and drop duplicates based on antecedent â†’ consequent
rules = rules.sort_values(by='confidence', ascending=False).drop_duplicates(subset=['antecedents', 'consequents'])
rules.head()
```

```python
# Drop Symmetric Rules
rules['pair_key'] = rules.apply(lambda row: tuple(sorted([row['antecedents'], row['consequents']])), axis=1)
rules = rules.drop_duplicates(subset='pair_key')
rules.drop(columns='pair_key', inplace=True)
```

```python
rules['antecedents'] = rules['antecedents'].apply(lambda x: list(x))
rules['consequents'] = rules['consequents'].apply(lambda x: list(x))
rules.to_csv("association_rules.csv", index=False)
```
